```python
import pandas as pd
import numpy as np

file_name = rf'/content/drive/My Drive/Datasets/accepted_2007_to_2018Q4 cleaned.csv'
loan_data_df = pd.read_csv(file_name)
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
loan_data_df['grade'] = label_encoder.fit_transform(loan_data_df['grade'])
loan_data_df['sub_grade'] = label_encoder.fit_transform(loan_data_df['sub_grade'])
loan_data_df['emp_length'] = label_encoder.fit_transform(loan_data_df['emp_length'])
loan_data_df['delinq_2yrs'] = label_encoder.fit_transform(loan_data_df['delinq_2yrs'])
loan_data_df['inq_last_12m'] = label_encoder.fit_transform(loan_data_df['inq_last_12m'])
loan_data_df['last_fico_range_low'] = label_encoder.fit_transform(loan_data_df['last_fico_range_low'])
loan_data_df['mort_acc'] = label_encoder.fit_transform(loan_data_df['mort_acc'])
loan_data_df['mths_since_last_delinq'] = label_encoder.fit_transform(loan_data_df['mths_since_last_delinq'])
loan_data_df['num_actv_bc_tl'] = label_encoder.fit_transform(loan_data_df['num_actv_bc_tl'])
loan_data_df['open_il_12m'] = label_encoder.fit_transform(loan_data_df['open_il_12m'])
loan_data_df['total_acc'] = label_encoder.fit_transform(loan_data_df['total_acc'])
loan_data_df.head()
```

**Binary Label Encoding**

```python
loan_data_df['term_'] = label_encoder.fit_transform(loan_data_df['term'])
loan_data_df['pymnt_plan'] = label_encoder.fit_transform(loan_data_df['pymnt_plan'])
loan_data_df['loan_status'] = label_encoder.fit_transform(loan_data_df['loan_status'])
loan_data_df.head()
```

**Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
loan_data_df['loan_amnt'] = scaler.fit_transform(loan_data_df[['loan_amnt']])
loan_data_df['days_since_issue_d'] = scaler.fit_transform(loan_data_df[['days_since_issue_d']])
loan_data_df['installment'] = scaler.fit_transform(loan_data_df[['installment']])
loan_data_df['int_rate'] = scaler.fit_transform(loan_data_df[['int_rate']])
loan_data_df['tot_cur_bal'] = scaler.fit_transform(loan_data_df[['tot_cur_bal']])
loan_data_df['total_rec_int'] = scaler.fit_transform(loan_data_df[['total_rec_int']])
loan_data_df['total_rec_late_fee'] = scaler.fit_transform(loan_data_df[['total_rec_late_fee']])
loan_data_df['total_rec_prncp'] = scaler.fit_transform(loan_data_df[['total_rec_prncp']])
loan_data_df['out_prncp'] = scaler.fit_transform(loan_data_df[['out_prncp']])
loan_data_df['annual_inc'] = scaler.fit_transform(loan_data_df[['annual_inc']])
loan_data_df['tot_hi_cred_lim'] = scaler.fit_transform(loan_data_df[['tot_hi_cred_lim']])
loan_data_df['days_since_last_cred_pull'] = scaler.fit_transform(loan_data_df[['days_since_last_cred_pull']])
loan_data_df.head()
```

**One Hot Encoding**

```python
purpose_encoded = pd.get_dummies(loan_data_df['purpose'], prefix='purpose')
home_ownership_encoded = pd.get_dummies(loan_data_df['home_ownership'], prefix='ownership')
loan_data_df = pd.concat([loan_data_df, purpose_encoded, home_ownership_encoded], axis=1)
loan_data_df.head()
```

**Target Encoding**

```python
column_to_encode_addr = 'addr_state'
mean_encoding_addr = loan_data_df.groupby(column_to_encode_addr)['loan_status_binary'].mean()
loan_data_df[f'{column_to_encode_addr}_encoded'] = loan_data_df[column_to_encode_addr].map(mean_encoding_addr)
column_to_encode_emp_title = 'emp_title'
mean_encoding_emp_title = loan_data_df.groupby(column_to_encode_emp_title)['loan_status_binary'].mean()
loan_data_df[f'{column_to_encode_emp_title}_encoded'] = loan_data_df[column_to_encode_emp_title].map(mean_encoding_emp_title)
loan_data_df.head()
```

**Exporting dataframe as csv file for modeling**

```python
loan_data_df.to_csv('/content/drive/My Drive/Datasets/accepted_2007_to_2018Q4 cleaned encoded.csv', index=False)
```
