```python
import pandas as pd
import matplotlib.pyplot as plt
import os

path = '/content/drive/My Drive/ML Project Fall 23'
os.chdir(path)
```

```python
import pandas as pd
file_name = rf'{path}/Datasets/accepted_2007_to_2018Q4.csv'
loan_data_df = pd.read_csv(file_name)
```

```python
# List of values to categorize
default_values = ['Charged Off', 'Default', 'Does not meet the credit policy. Status:Charged Off', 'Late (31-120 days)', 'In Grace Period']
# Create a new column 'category' based on your criteria
loan_data_df['category'] = loan_data_df['loan_status'].apply(lambda x: 'default' if x in default_values else 'non-default')

loan_data_df.head(20)

loan_data_updated_df = loan_data_df[["id","purpose","loan_amnt","term","issue_d","installment","int_rate","grade","sub_grade",
                                     "tot_cur_bal","pymnt_plan","addr_state","total_rec_int","total_rec_late_fee",
                                     "total_rec_prncp","out_prncp","emp_title","emp_length","annual_inc","home_ownership",
                                     "delinq_2yrs","inq_last_12m","last_credit_pull_d","last_fico_range_low","mort_acc",
                                     "mths_since_last_delinq","num_actv_bc_tl","open_il_12m","tot_hi_cred_lim","total_acc",
                                     "loan_status","category"]]
loan_data_updated_df.head(10)

loan_data_updated_df.isnull().sum()

nan_values_purpose = loan_data_updated_df[loan_data_updated_df['purpose'].isna()]
print (nan_values_purpose)

loan_data_updated_df.dropna(subset=["purpose"],inplace=True)

loan_data_updated_df.isnull().sum()

loan_data_updated_df.dtypes

loan_data_updated_df["purpose"].unique()
```

```python
# Categorizing data in 'purpose' field
purpose_house = ['house','home_improvement','major_purchase','moving', 'renewable_energy']
purpose_other = ['other','wedding','educational','medical','small_business','vacation']
# Define the categorization function
def categorize_purpose (purpose):
    if purpose in purpose_house:
        return 'house'
    elif purpose in purpose_other:
        return 'other'
    else:
        return purpose
# Apply the categorization function to create a new field
loan_data_updated_df['purpose_new']=loan_data_updated_df['purpose'].apply(categorize_purpose)

loan_data_updated_df['purpose_new'].unique()
```

```python

total_acc_bins = [0, 10, 20, 50, 100, float('inf')]
total_acc_labels = ['0-10', '10-20', '20-50', '50-100', '100+']

loan_data_updated_df['total_acc_new'] = pd.cut(loan_data_updated_df['total_acc'],
                                               bins=total_acc_bins, labels=total_acc_labels, right=False)

loan_data_updated_df[['total_acc','total_acc_new']].head(10)

loan_data_updated_df['tot_hi_cred_lim_new'] = loan_data_updated_df['tot_hi_cred_lim']
loan_data_updated_df['tot_hi_cred_lim_new'].fillna(0, inplace=True)

null_value = loan_data_updated_df[loan_data_updated_df['tot_hi_cred_lim'].isnull()][['tot_hi_cred_lim','tot_hi_cred_lim_new']]
null_value.head(10)

loan_data_updated_df['open_il_12m'].fillna(0, inplace=True)
open_il_12m_bins = [0, 2, 5, float('inf')]
open_il_12m_labels = ['0-2', '2-5', '5+']
loan_data_updated_df['open_il_12m_new'] = pd.cut(loan_data_updated_df['open_il_12m'],
                                                 bins=open_il_12m_bins, labels=open_il_12m_labels, right=False)

loan_data_updated_df[['open_il_12m','open_il_12m_new']].head(10)

loan_data_updated_df['num_actv_bc_tl'].fillna(0, inplace=True)
num_actv_bc_tl_bins = [0, 2, 5,10,20,float('inf')]
num_actv_bc_tl_labels = ['0-2', '2-5','5-10','10-20','20+']

loan_data_updated_df['num_actv_bc_tl_new'] = pd.cut(loan_data_updated_df['num_actv_bc_tl'], bins=num_actv_bc_tl_bins,
                                                    labels=num_actv_bc_tl_labels, right=False)

loan_data_updated_df[['num_actv_bc_tl','num_actv_bc_tl_new']].head(10)
```

```python

# Checking for missing data in 'annual_inc' field
nan_values_annual_inc = loan_data_updated_df[loan_data_updated_df['annual_inc'].isna()]
print (nan_values_annual_inc)

# Dropping null values in 'annual_inc' field
loan_data_updated_df.dropna(subset=["annual_inc"],inplace=True)

nan_values_annual_inc_check = loan_data_updated_df[loan_data_updated_df['annual_inc'].isna()]
print(nan_values_annual_inc_check)

# Checking for different type of values or categories present in 'home_ownership' field
loan_data_updated_df['home_ownership'].unique()

# Dropping the rows or the instances where there value of 'home_ownership' is 'ANY'
loan_data_updated_df=loan_data_updated_df[loan_data_updated_df['home_ownership']!="ANY"]

# Re-checking for different type of values or categories present in 'home_ownership' field
loan_data_updated_df['home_ownership'].unique()

loan_data_updated_df['num_actv_bc_tl'].fillna(0, inplace=True)

num_actv_bc_t1_bins = [0, 2, 5, 10, 20, float('inf')]
num_actv_bc_t1_labels = ['0-2', '2-5','5-10','10-20', '20+']
loan_data_updated_df['num_actv_bc_tl_new'] = pd.cut(loan_data_updated_df['num_actv_bc_tl'],
                                                    bins=num_actv_bc_t1_bins, labels=num_actv_bc_t1_labels, right=False)

loan_data_updated_df[['num_actv_bc_tl','num_actv_bc_tl_new']].head(10)
```

```python

# Checking for missing data in 'delinq_2yrs' field
nan_values_delinq_2yrs = loan_data_updated_df[loan_data_updated_df['delinq_2yrs'].isna()]
print (nan_values_delinq_2yrs)

# Dropping null values in 'delinq 2yrs' field
loan_data_updated_df.dropna (subset=["delinq_2yrs"], inplace=True)

nan_values_delinq_2yrs = loan_data_updated_df[loan_data_updated_df['annual_inc'].isna()]
print (nan_values_delinq_2yrs)


# Binning and providing appropriate labels for the data present in 'delinq_2yrs' field
delinq_2yrs_bins = [0,2,5,10,float('inf')]
delinq_2yrs_labels = ['0-2', '2-5', '5-10','10+']
loan_data_updated_df['delinq_2yrs_new'] = pd.cut(loan_data_updated_df['delinq_2yrs'],
                                                  bins=delinq_2yrs_bins, labels=delinq_2yrs_labels, right=False)
loan_data_updated_df[['delinq_2yrs', 'delinq_2yrs_new']].head(10)
```

```python

# Creating a new field by binning and providing appropriate labels for the data present in 'inq_last_12m' field
loan_data_updated_df['inq_last_12m'].fillna(0, inplace=True)
inq_last_12m_bins = [0, 2, 5,10,20,float('inf')]
inq_last_12m_labels = ['0-2', '2-5', '5-10', '10-20','20+']
loan_data_updated_df['inq_last_12m_new'] = pd.cut(loan_data_updated_df['inq_last_12m'],
                                                   bins=inq_last_12m_bins, labels=inq_last_12m_labels, right=False)
loan_data_updated_df[['inq_last_12m', 'inq_last_12m_new']].head(10)

mths_since_last_delinq_bins = [0,20,50,float('inf')]
mths_since_last_delinq_labels = ['0-20', '20-50', '50+']
# Creating the 'mths_since_last_delinq_category' colunn
loan_data_updated_df['mths_since_last_delinq_category'] = pd.cut(loan_data_updated_df['mths_since_last_delinq'],
                                                                bins=mths_since_last_delinq_bins,
                                                                labels=mths_since_last_delinq_labels, right=False)

# Define a mapping of categories
category_mapping = {
    '0-20': 'Very Recent',
    '20-50': 'Recent',
    '50+': 'Old'
}

loan_data_updated_df['mths_since_last_delinq_new'] = loan_data_updated_df['mths_since_last_delinq_category'].replace(category_mapping)

# Add 'Never' as a new category
loan_data_updated_df['mths_since_last_delinq_new']=loan_data_updated_df['mths_since_last_delinq_new'].astype('category')
loan_data_updated_df['mths_since_last_delinq_new']=loan_data_updated_df['mths_since_last_delinq_new'].cat.add_categories('Never')

# Fill 'Never' for NaN values
loan_data_updated_df['mths_since_last_delinq_new'] = loan_data_updated_df['mths_since_last_delinq_new'].fillna('Never')

loan_data_updated_df [['mths_since_last_delinq', 'mths_since_last_delinq_new']].head(10)
```

```python

# Creating a new field by binning and providing appropriate labels for the data present in 'last_fico_range_low' field
last_fico_range_low_bins = [0, 500, 600, 700, 800, float('inf')]
last_fico_range_low_labels = ['No Credit', 'Poor', 'Average', 'Good', 'Excellent']
loan_data_updated_df['last_fico_range_low_new'] = pd.cut (loan_data_updated_df['last_fico_range_low'],
                                                          bins=last_fico_range_low_bins,
                                                          labels=last_fico_range_low_labels,right=False)
loan_data_updated_df[['last_fico_range_low', 'last_fico_range_low_new']].head(10)

# Creating a new field by binning and providing appropriate labels for the data present in 'mort_acc' field
loan_data_updated_df['mort_acc'].fillna(0, inplace=True)
mort_acc_bins = [0,2,5,10,float('inf')]
mort_acc_labels = ['0-2', '2-5', '5-10', '10+']
loan_data_updated_df['mort_acc_new'] = pd.cut (loan_data_updated_df['mort_acc'],
                                               bins=mort_acc_bins, labels=mort_acc_labels, right=False)
loan_data_updated_df[['mort_acc', 'mort_acc_new']].head(10)

nan_values_last_credit_pull_d = loan_data_updated_df[loan_data_updated_df['last_credit_pull_d'].isna()]
print(nan_values_last_credit_pull_d)

# Dropping null values in 'last_credit_pull_d' field
loan_data_updated_df.dropna(subset=["last_credit_pull_d"],inplace=True)
```

```python

######################################################################################################
#Clean emp_title column

# Modifications for "emp_title" and "annual_inc" columns
def update_emp_title(row):
    if pd.isna(row['emp_title']):
        if row['annual_inc'] > 0:
            return 'missing'
        elif row['annual_inc'] == 0:
            return 'unemployed'
    return row['emp_title']

loan_data_updated_df['emp_title'] = loan_data_updated_df.apply(update_emp_title, axis=1)
loan_data_updated_df = loan_data_updated_df.dropna(subset=['emp_title'])
```

```python
######################################################################################################
#Clean emp_length column

# Extracting integer value from "emp_length" column
loan_data_updated_df['emp_length'] = loan_data_updated_df['emp_length'].str.extract('(\d+)').astype(float)

# Handling nan values in "emp_length" based on "annual_inc" values
def handle_nan(row):
    if pd.isna(row['emp_length']) and row['annual_inc'] == 0:
        return 0.0  # Ensure this is a float to match the other data
    return row['emp_length']

loan_data_updated_df['emp_length'] = loan_data_updated_df.apply(handle_nan, axis=1)

# Binning the data in "emp_length" column
bins = [0, 1, 4, 9, float('inf')]
labels = ['unemployed', 'short', 'mid', 'long']
loan_data_updated_df['emp_length'] = pd.cut(loan_data_updated_df['emp_length'], bins=bins, labels=labels, right=False, include_lowest=True)

# Handling remaining nan values as 'missing'
loan_data_updated_df['emp_length'] = loan_data_updated_df['emp_length'].cat.add_categories('missing').fillna('missing')
```

```python
######################################################################################################
#Clean total_cur_bal column
def handle_tot_cur_bal(row):
    if pd.isna(row['tot_cur_bal']):
        if row['loan_status'] == 'non-default':
            return 0.0  # Ensure this is a float to match the other data
        elif row['loan_status'] == 'default':
            return pd.NA  # Use pd.NA to mark rows for deletion
    return row['tot_cur_bal']

# Apply the function to each row
loan_data_updated_df['tot_cur_bal'] = loan_data_updated_df.apply(handle_tot_cur_bal, axis=1)

# Remove rows marked for deletion
loan_data_updated_df = loan_data_updated_df[loan_data_updated_df['tot_cur_bal'].notna()]
```

```python
######################################################################################################
#Add derived features days_since_issue_d and days_since_last_cred_pull
#days_since_issue_d: todays date - issue_d, total days as integer
#days_since_last_cred_pull: todays date - last_credit_pull_d, total days as integer

#Convert the date strings in 'issue_d' and 'last_credit_pull_d' to datetime objects
loan_data_updated_df['issue_d'] = pd.to_datetime(loan_data_updated_df['issue_d'], format='%b-%Y')
loan_data_updated_df['last_credit_pull_d'] = pd.to_datetime(loan_data_updated_df['last_credit_pull_d'], format='%b-%Y')

from datetime import datetime
# Get today's date
today = datetime.now()

# Calculate the number of days from 'issue_d' and 'last_credit_pull_d' to today
loan_data_updated_df['days_since_issue_d'] = (today - loan_data_updated_df['issue_d']).dt.days
loan_data_updated_df['days_since_last_cred_pull'] = (today - loan_data_updated_df['last_credit_pull_d']).dt.days
```

```python
#######################################################################################################
#Get rid of "_new" in the new columns. This will get rid of the non-cleaned original columns

# Get a list of original column names to be dropped and a dictionary for renaming
columns_to_drop = [col.replace('_new', '') for col in loan_data_updated_df.columns if '_new' in col]
rename_dict = {col: col.replace('_new', '') for col in loan_data_updated_df.columns if '_new' in col}

# Drop the original columns
loan_data_updated_df = loan_data_updated_df.drop(columns=columns_to_drop)
loan_data_updated_df = loan_data_updated_df.rename(columns=rename_dict)


#Drop "loan_status" column
loan_data_updated_df = loan_data_updated_df.drop(columns = ['loan_status'])

#Rename "category" to loan_status"
loan_data_updated_df = loan_data_updated_df.rename(columns={'category': 'loan_status'})

print(loan_data_updated_df.isnull().sum())

new_file_name = file_name.replace(".csv", " cleaned.csv")
loan_data_updated_df.to_csv(new_file_name, index=False)

#Get top 100000 rows
loan_data_df_sample = loan_data_updated_df.head(100000)
new_file_name = file_name.replace(".csv", " cleaned top 100000 rows.csv")
loan_data_df_sample.to_csv(new_file_name, index=False)
```
